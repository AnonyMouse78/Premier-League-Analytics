```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO
import time
```

```python
#Retrieve HTML data & store as python object
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
data = requests.get(standings_url)
#To describe the raw HTML data:
#data.text
```

```python
#Initialize soup object
soup = BeautifulSoup(data.text)
standings_table = soup.select('table.stats_table')[0]  #select uses CSS selectors (elements, classes, id)
#print(standings_table.prettify())  --To view standings table narrowed down from the vast data pool!--
links = standings_table.find_all('a')  #find_all only finds tags (simpler but replacable)
#find href's for every link
links = [l.get("href") for l in links]
#find only links of squads ie. having "/squads/" in the href & get rid of rest
links = [l for l in links if '/squads/' in l]
links
```

```python
#python string formatting to complete the urls
team_urls = [f"https://fbref.com{l}" for l in links]
team_urls
```

```python
# Scrape data for multiple teams over multiple years

years = list(range(2022, 2020, -1))
all_matches = []
standings_url = "https://fbref.com/en/comps/9/Premier-League-Stats"
```

```python
for year in years:
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text)
    standings_table = soup.select('table.stats_table')[0] #Individual match data for each team
    
    links = [l.get("href") for l in standings_table.find_all('a')] #Get all hrefs from standings
    links = [l for l in links if '/squads/' in l] #Filter only squads info
    team_urls = [f"https://fbref.com{l}" for l in links] #Convert relative to abs links
    
    #Scrape the previous season data:
    previous_season = soup.select("a.prev")[0].get("href")  #Select a tags with prev & get href of anchor tags
    standings_url = f"https://fbref.com{previous_season}"  #Convert to abs URL & when loop continues, gets standings_url or prev season
    
    for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("Stats", "").replace("-","")
        
        data = requests.get(team_url)
        matches = pd.read_html(StringIO(data.text), match="Scores & Fixtures")[0]
        
        soup = BeautifulSoup(data.text)
        links = [l.get("href") for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]  #Filter only 'all_comps/shooting/' & link
        data = requests.get(f"https://fbref.com{links[0]}")  #Relative to abs link
        shooting = pd.read_html(StringIO(data.text), match="Shooting")[0]
        shooting.columns = shooting.columns.droplevel()
        
        #Some teams dont have shooting stats, handle ValueError
        
        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
        except ValueError:
            continue
        
        #Add some additional data to team_data -> No team name & season inside table
        team_data = team_data[team_data["Comp"] == "Premier League"]
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        #Sleeps for '1 or 5 or 10' seconds to prevent excess scaping // Avoid Table not found ValueError
        time.sleep(10)
```
